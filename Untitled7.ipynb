```python
from collections import Counter
import pandas as pd
age = ["youth", "youth", "middle_aged", "senior", "senior", "senior", "middle_aged", "youth", "youth", 
       "senior", "youth", "middle_aged", "middle_aged", "senior"]

income = ["high", "high", "high", "medium", "low", "low", "low", "medium", "low", 
          "medium", "medium", "medium", "high", "medium"]

student = ["no", "no", "no", "no", "yes", "yes", "yes", "no", "yes", 
           "yes", "yes", "no", "yes", "no"]

credit_rating = ["fair", "excellent", "fair", "fair", "fair", "excellent", "excellent", "fair", "fair", 
                 "fair", "excellent", "excellent", "fair", "excellent"]

buy_computer = ["no", "no", "yes", "yes", "yes", "no", "yes", "no", "yes", 
                "yes", "yes", "yes", "yes", "no"]
data = {
    "age": age,
    "income": income,
    "student": student,
    "credit_rating": credit_rating,
    "class": buy_computer
}
df = pd.DataFrame(data)

class_counts = Counter(df["class"])
total_instances = len(df)

prior_yes = class_counts["yes"] / total_instances
prior_no = class_counts["no"] / total_instances

def conditional_prob(feature, value, target_class):
    subset = df[df["class"] == target_class]
    return len(subset[subset[feature] == value]) / len(subset) if len(subset) > 0 else 0
features = {
    "age": "youth",
    "income": "medium",
    "student": "yes",
    "credit_rating": "fair"
}
likelihood_yes = 1
likelihood_no = 1

for feature, value in features.items():
    likelihood_yes *= conditional_prob(feature, value, "yes")
    likelihood_no *= conditional_prob(feature, value, "no")

# Calculate posterior probabilities
posterior_yes = likelihood_yes * prior_yes
posterior_no = likelihood_no * prior_no

print("Posterior Probability (yes):", posterior_yes)
print("Posterior Probability (no):", posterior_no)

if posterior_yes > posterior_no:
    print("Predicted Class: yes")
else:
    print("Predicted Class: no")
```
